```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/heart.csv')
df.head()
```

```python
df.info()
```

```python
sex_counts = df['sex'].value_counts()
print(f"Female: {sex_counts.get(0, 0)}")
print(f"Male: {sex_counts.get(1, 0)}")
```

```python

import matplotlib.pyplot as plt
male_count = df[df['sex'] == 1].shape[0]
female_count = df[df['sex'] == 0].shape[0]
labels = ['Male', 'Female']
sizes = [male_count, female_count]
colors = ['black', 'pink']
explode = (0.1, 0)  # explode 1st slice (Male)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Male vs Female Distribution')

plt.show()
```

```python
import altair as alt

chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('sex:N', title='Sex (0: Female, 1: Male)'),
    y=alt.Y('count()', title='Count')
)

chart
```

```python

import matplotlib.pyplot as plt

cp_counts = df['cp'].value_counts()

plt.pie(cp_counts, labels=cp_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Chest Pain Types')
plt.show()
```

```python
import altair as alt

chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('cp', bin=alt.Bin(extent=[0, 3], step=1)),
    y='count()'
)

chart
```

```python
# prompt: make a pie bar graph for exercise induced angina (exang) values are 1 means yes 0 means no

import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df' and contains the 'exang' column

exang_counts = df['exang'].value_counts()

# Pie chart
plt.pie(exang_counts, labels=['No (0)', 'Yes (1)'], autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Ensure a circular pie chart
plt.title('Exercise Induced Angina')
plt.show()

# Bar chart
plt.bar(exang_counts.index, exang_counts.values)
plt.xlabel('Exercise Induced Angina (0: No, 1: Yes)')
plt.ylabel('Count')
plt.title('Distribution of Exercise Induced Angina')
plt.xticks([0, 1])
plt.show()
```

```python

sex_target_counts = df.groupby(['sex', 'target'])['target'].count().unstack()
sex_target_counts.plot(kind='bar', stacked=True)
plt.xlabel('Sex (0: Female, 1: Male)')
plt.ylabel('Count')
plt.title('Sex vs. Target (Heart Disease)')
plt.show()



cp_target_counts = df.groupby(['cp', 'target'])['target'].count().unstack()
cp_target_counts.plot(kind='bar', stacked=True)
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.title('Chest Pain Type vs. Target (Heart Disease)')
plt.show()


fbs_target_counts = df.groupby(['fbs', 'target'])['target'].count().unstack()
fbs_target_counts.plot(kind='bar', stacked=True)
plt.xlabel('Fasting Blood Sugar (0: False, 1: True)')
plt.ylabel('Count')
plt.title('Fasting Blood Sugar vs. Target (Heart Disease)')
plt.show()
```
